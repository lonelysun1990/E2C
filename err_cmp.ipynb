```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
plt.set_cmap('jet')
```

```python
# data_dir = '/data3/Astro/lstm_rom/e2c_larry/data/9w_bhp_rate_fix_wl/'
data_dir = '/data3/Astro/lstm_rom/e2c_larry/data/9w_ms_bhp_rate_fix_wl/'

case_name = '9w_ms_bhp_rate'

target_suffix = '_fix_wl_rel_8' # the dataset being evaluated here
train_suffix = '_with_p'
model_suffix = '_flux_loss'

num_t = 20
run_eval = 100
num_eval = num_t * run_eval
# dt = 200 // num_t
dt = 100 # days
```

```python
nt = 6000
hf_r = h5py.File(data_dir + case_name + target_suffix + '_nt_%d_pred_err.mat'%(nt), 'r')
pres_rel_err_true_pred_1 = np.array(hf_r.get('pres_rel_err_true_pred'))
pres_rel_err_true_close_train_1 = np.array(hf_r.get('pres_rel_err_true_close_train'))
sat_rel_err_true_pred_1 = np.array(hf_r.get('sat_rel_err_true_pred'))
sat_rel_err_true_close_train_1 = np.array(hf_r.get('sat_rel_err_true_close_train'))
inj_bhp_rel_err_true_pred_1 = np.array(hf_r.get('inj_bhp_rel_err_true_pred'))
inj_bhp_rel_err_true_close_train_1 = np.array(hf_r.get('inj_bhp_rel_err_true_close_train'))
rates_rel_err_true_pred_1 = np.array(hf_r.get('rates_rel_err_true_pred'))
rates_rel_err_true_close_train_1 = np.array(hf_r.get('rates_rel_err_true_close_train'))

hf_r.close()
```
